```python
# 로또 번호는 독립시행이라 딥러닝이 의미가 없을테지만
# 그래도 한번 분석한 번호라는 기분이라도 내보려고 만들어 보게 되었다.

# 당첨번호 데이터는 https://dhlottery.co.kr/gameResult.do?method=byWin&wiselog=H_C_1_1 에서 받음
```

```python
import numpy as np
import pandas as pd

past_nums = pd.read_csv('./lotto.csv')
```

```python
past_nums.insert(2, 'year', past_nums['date'])
past_nums.insert(3, 'month', past_nums['date'])
past_nums.insert(4, 'day', past_nums['date'])
past_nums.drop('date', axis=1, inplace=True)

# 2등은 노리지 않아서 보너스 번호를 없앨 경우
past_nums.drop('bonus', axis=1, inplace=True)
```

```python
def get_year(x):
    return int(x.split('.')[0])
def get_month(x):
    return int(x.split('.')[1])
def get_day(x):
    return int(x.split('.')[2])
```

```python
past_nums['year'] = past_nums['year'].apply(get_year)
past_nums['month'] = past_nums['month'].apply(get_month)
past_nums['day'] = past_nums['day'].apply(get_day)
```

```python
past_nums
```

```python
def year_onehot(year):

    ohbin = np.zeros(22)
    ohbin[int(year)-2002] = 1
    
    return ohbin
def month_onehot(month):

    ohbin = np.zeros(12)
    ohbin[int(month)-1] = 1
    
    return ohbin
def day_onehot(day):

    ohbin = np.zeros(31)
    ohbin[int(day)-1] = 1
    
    return ohbin

def wn_onehot(numbers):

    ohbin = np.zeros(45) #45개의 빈 칸을 만듬

    for i in range(6): #여섯개의 당첨번호에 대해서 반복함
        ohbin[int(numbers[i])-1] = 1 #로또번호가 1부터 시작하지만 벡터의 인덱스 시작은 0부터 시작하므로 1을 뺌
    
    return ohbin
```

```python
# year one-hot
years = past_nums.loc[:, 'year']
years = years.to_numpy()
ohbins = list(map(year_onehot, years))

past_nums.drop('year', axis=1, inplace=True)
past_nums.insert(1, 'year', ohbins)
```

```python
# month one-hot
month = past_nums.loc[:, 'month']
month = month.to_numpy()
ohbins = list(map(month_onehot, month))

past_nums.drop('month', axis=1, inplace=True)
past_nums.insert(2, 'month', ohbins)
```

```python
# day one-hot
day = past_nums.loc[:, 'day']
day = day.to_numpy()
ohbins = list(map(day_onehot, day))

past_nums.drop('day', axis=1, inplace=True)
past_nums.insert(3, 'day', ohbins)
```

```python
# 테스트 인덱스 삭제
past_nums.drop(1063, inplace=True)
```

```python
numbers = past_nums.loc[:, 'wn1':'wn6']
numbers = numbers.to_numpy()
ohbins = list(map(wn_onehot, numbers))

past_nums.insert(len(past_nums.columns), 'wn', ohbins)
past_nums.drop(['wn1', 'wn2', 'wn3', 'wn4', 'wn5', 'wn6'], axis=1, inplace=True)
```

```python
past_nums.drop('round', axis=1, inplace=True)
```
